#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def task1 (a: str) -> int:
    vowels = {"a", "e", "i", "o", "u"}
    a = a.lower()
    count = 0
    for char in a:
        if char in vowels:
            count += 1

    return count

assert task1("cat") == 1
assert task1("pictures") == 3
assert task1("beautiful") == 5
assert task1("UniversIty") == 4
assert task1("green") == 2 
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def task2(s: str) -> bool:

    text = set()
    
    for char in s:
        if char in text:
            return False
          
        text.add(char)
    
    return True  

assert task2("aa") == False
assert task2("ab") == True
assert task2("abc") == True
assert task2("aabc") == False
assert task2("") == True
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def task3(n: int) -> int:
    binary = bin(n)
    return binary.count("1")


assert task3(0) == 0
assert task3(1) == 1
assert task3(5) == 2   # 101
assert task3(7) == 3  # 111
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(vector1, vector2):
    if len(vector1) != len(vector2):
        raise ValueError("векторы должны быть одной длины")

    squared_errors = []

    for i in range(len(vector1)):
        diff = vector1[i] - vector2[i]
        squared = diff * diff
        squared_errors.append(squared)

    total_sum = sum(squared_errors)
    return total_sum / len(squared_errors)


# Тесты
assert mse([1, 2, 3], [1, 2, 3]) == 0
assert mse([1, 2, 3], [2, 3, 4]) == 1
assert abs(mse([1, 2], [2, 4]) - 2.5) < 1e-9
assert mse([0, 0, 0], [1, 1, 1]) == 1
assert mse([-1, -2, -3], [-1, -2, -4]) == (0 + 0 + 1) / 3

try:
    mse([1, 2], [1])
    raise AssertionError("должна быть ошибка при разной длине векторов")
except ValueError:
    pass

```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def task6(n: int) -> str:
    factors = []
    divisor = 2

    while divisor * divisor <= n:
        count = 0
        while n % divisor == 0:
            n = n // divisor
            count += 1

        if count == 1:
            factors.append(f"({divisor})")
        elif count > 1:
            factors.append(f"({divisor}**{count})")

        divisor += 1

    if n > 1:
        factors.append(f"({n})")

    return "".join(factors)


assert task6(4) == "(2**2)"
assert task6(86240) == "(2**5)(5)(7**2)(11)"
assert task6(13) == "(13)"
assert task6(128) == "(2**7)"
assert task6(832) == "(2**6)(13)"

        
    
    

    
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def task7(blocks):
    sum_blocks = 0
    level = 0

    while sum_blocks < blocks:
        level += 1
        sum_blocks += level * level

    return level if sum_blocks == blocks else "It is impossible"


assert task7(5) == 2
assert task7(30) == 4
assert task7(91) == 6
assert task7(385) == 10
assert task7(50) == "It is impossible"
assert task7(4) == "It is impossible"
assert task7(141) == "It is impossible"
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def task8(number: int) -> bool:
    digits = str(number)
    length = len(digits)

    if length % 2 == 0:
        first_part = digits[:length // 2 - 1]
        second_part = digits[length // 2 + 1:]
    else:
        first_part = digits[:length // 2]
        second_part = digits[length // 2 + 1:]

    sum_left = sum(int(d) for d in first_part)
    sum_right = sum(int(d) for d in second_part)
    
    return sum_left == sum_right

assert task8(121) == True
assert task8(1238033) == True
assert task8(1234003) == False
assert task8(12347204) == True
assert task8(12301214) == False
assert task8(11) == True  # 1|1
assert task8(1221) == True  # 1|2 2|1 -> 1 == 1
assert task8(1234) == False  # 1|3 4| -> 1 != 0
```
